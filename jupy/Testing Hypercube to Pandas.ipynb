```python
import pandas as pd
import os
from matplotlib import pyplot as plt
from spectral import *
```

```python
PROJ_DIR = os.path.dirname(os.getcwd())
SAMPLES_DIR = os.path.join(PROJ_DIR, 'samples')
SAMPLE_FILE = SAMPLES_DIR + '\\HARC000.bil.hdr'
SAMPLE_FILE
```

### Abrimos la imágen hiperspectral

```python
SAMPLE = open_image(SAMPLE_FILE)
SAMPLE
```

### Cargamos el conjunto de datos de la Imágen en Memoria 

```python
s = SAMPLE.load()
```

### Convertimos el hipercubo en un arreglo de datos Numpy

```python
a = s.asarray()
```

### Ploteamos la firma espectral del primer Píxel

```python
for pixel in range(10):
    plt.plot(a[0,0+pixel,:])
plt.xlabel('Banda')
plt.ylabel('Reflectancia del Pixel')
plt.show()
```
